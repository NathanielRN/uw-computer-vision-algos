## Problem 1
### Prove that affine transformations map lines onto lines. For this, take an arbitrary line in $R^2$ and show that an arbitrary affine transofrmation maps it onto a set of points that satisfy a line equation. Use linear algebraic equations for lines, i.e. $x^Tv=c$ where $v$ is a 2-vector (normal to the line), $c$ is a scalar ($v,c$ are line parameters), and $x$ is a 2-vector corresponding to an arbitrary point on the line. Your proof should use only linear algebraic equations. 
### HINT: for inspiration, check out the proof for homographies on slide 33, topic 4.

Solution:

Take an arbitrary line in $R^2$, $y = ax + b$. The points on this line are $p = (x,y)$. We can define a vector $\vec{p} = \left[\begin{array}{cc} x \\ y \\ 1 \end{array}\right]$ to represet all points in homogeneous coordinates for convenience. The equation of the line can be rewritten in matrix notation as follows:

$$\begin{aligned}
\left[
\begin{array}{cc}
-a & 1 & -b \\
\end{array}
\right] \cdot \left[
\begin{array}{cc}
x \\
y \\
1 \\
\end{array}
\right]
&= 0 \\
\vec{c}^T \cdot \vec{p} &= 0
\end{aligned}$$

An affine transformation matrix has the following form:

$$\left[
\begin{array}{cc}
a & b & c \\
d & e & f \\
0 & 0 & 1 \\
\end{array}
\right]$$

where $a$, $b$, $c$, $d$, $e$, $f$ are scalars. Let this matrix be called $A$.

It has a determinant as follows:

$$\begin{aligned}
det(A) &= det\left(\left[
\begin{array}{cc}
a & b & c \\
d & e & f \\
0 & 0 & 1 \\
\end{array}
\right]\right)
&= (1) \cdot \left|ae - bd\right|
\end{aligned}$$

As long as $ae \neq bd$ the determinant of $A$, $det(A) \neq 0$ and the inverse matrix $A^{-1}$ is defined.

We know that an affine transformation takes points $(x, y)$ and converts them to new points $(x', y')$. This can be represented using matrix multiplication and their homogenous coordinate representations.

$$\begin{aligned}
\left[
\begin{array}{cc}
a & b & c \\
d & e & f \\
0 & 0 & 1 \\
\end{array}
\right] \cdot \left[
\begin{array}{cc}
x \\
y \\
1 \\
\end{array}
\right]
&= \left[
\begin{array}{cc}
x' \\
y' \\
w' \\
\end{array}
\right] \\
A \cdot \left[
\begin{array}{cc}
x \\
y \\
1 \\
\end{array}
\right]
&= \left[
\begin{array}{cc}
x' \\
y' \\
w' \\
\end{array}
\right]
\end{aligned}$$

These new points $p' = (\frac{x'}{w'}, \frac{y'}{w'})$ form a line like they did for the points $p$, $\frac{y'}{w'} = a'\frac{x'}{w'} + b'$ or $y' = a'x' + b'w'$. In matrix notation and using homogenous coordinates $\vec{p} = \left[\begin{array}{cc} x' \\ y' \\ w' \end{array}\right]$:

$$\begin{aligned}
\left[
\begin{array}{cc}
-a' & 1 & -b' \\
\end{array}
\right] \cdot \left[
\begin{array}{cc}
x' \\
y' \\
w' \\
\end{array}
\right]
&= 0 \\
\vec{c'}^T \cdot \vec{p'} &= 0
\end{aligned}$$

However, since we know the transformation that was used to obtain the points $p' = (\frac{x'}{w'}, \frac{y'}{w'})$, we can substitute that back into the line equation for the new points to get a line equation using the original points:

$$\begin{aligned}
\vec{c'}^T \cdot \vec{p'} &= 0 \\
\vec{c'}^T \cdot \left[
\begin{array}{cc}
x' \\
y' \\
w' \\
\end{array}
\right] &= 0 \\
\vec{c'}^T \cdot \left(A \cdot \left[
\begin{array}{cc}
x \\
y \\
1 \\
\end{array}
\right]\right) &= 0 \\
\left(\vec{c'}^T \cdot A\right) \cdot \left[
\begin{array}{cc}
x \\
y \\
1 \\
\end{array}
\right] &= 0 \\
\left(\vec{c'}^T \cdot A\right) \cdot \vec{p} &= 0
\end{aligned}$$

This looks very similar to the matrix notation for the original arbitrary line $y = ax + b$.

We want a non-trivial solution such that this equation is still true but $\vec{p} \neq \vec{0}$. Comparing this with the equation $\vec{c}^T \cdot \vec{p} = 0$ from above, we try to substitue $\vec{c'}^T = \vec{c}^T \cdot \left(A^{-1}\right)$.

The points $p' = (\frac{x'}{w'}, \frac{y'}{w'})$ form a unique line. This means the vector $\vec{c'}$ to make that line is also unique. Finally, the substitution $\vec{c'}^T = \vec{c}^T \cdot \left(A^{-1}\right)$ must as a consequence also be a unqiue solution.

Using the substitution here:

$$\begin{aligned}
\vec{c'}^T \cdot A \cdot \vec{p} &= 0 \\
(\vec{c}^T \cdot A^{-1}) \cdot A \cdot \vec{p} &= 0 \\
\vec{c}^T \cdot \left(A^{-1} \cdot A\right) \cdot \vec{p} &= 0 \\
\vec{c}^T \cdot \vec{p} &= 0 \\
\left[
\begin{array}{cc}
-a & 1 & -b \\
\end{array}
\right] \cdot \left[
\begin{array}{cc}
x \\
y \\
1 \\
\end{array}
\right]
&= 0
\end{aligned}$$

We notice that the resulting line $y' = a'x' + b'w'$ defines the same line originally defined by $y = ax + b$.

This means that for affine transformation matrix $A$ with $ae \neq bd$ and $\vec{c'} = \left(A^{-1}\right)^T \cdot \vec{c}$, the new line $y' = a'x' + b'w'$ is equal to the original line $y = ax + b$ meaning lines map to lines.

## Problem 2
### Prove that any roto-reflective transformation $R^2 \rightarrow R^2$ defined by 2x2 orthogonal matrix $R$ (s.t. $R^TR=I$) preserves (a) orthogonal lines and (b) distances between points. Your proof should use only linear algebraic equations. 

Solution:

For part a:

Take some vector $\left[\begin{array}{cc} x_1 \\ y_1 \end{array}\right]$ and and another vector that we know is orthogonal to it $\left[\begin{array}{cc} x_2 \\ y_2 \end{array}\right]$ both in $R^2$.

This means the scalar product between them is $0$. i.e.:

$$\begin{aligned}
\left[
\begin{array}{cc}
x_1 \\
y_1 \\
\end{array}
\right] \cdot
\left[
\begin{array}{cc}
x_2 \\
y_2 \\
\end{array}
\right] &= 0 \\
\left(\left[
\begin{array}{cc}
x_1 \\
y_1 \\
\end{array}
\right]\right)^T
\left[
\begin{array}{cc}
x_2 \\
y_2 \\
\end{array}
\right] &= 0 \\
\left[
\begin{array}{cc}
x_1 & y_1
\end{array} 
\right]
\left[ \begin{array}{cc}
x_2 \\
y_2 \\
\end{array}
\right] &= 0 \\
\end{aligned}
$$

The transformation of these vectors are as follows:

$$R \cdot \left[\begin{array}{cc}
x_1 \\
y_1 \\
\end{array}
\right]= \left[\begin{array}{cc}
x_1' \\
y_1' \\
\end{array}
\right]
$$

$$R \cdot \left[\begin{array}{cc}
x_2 \\
y_2 \\
\end{array}
\right]= \left[\begin{array}{cc}
x_2' \\
y_2' \\
\end{array}
\right]
$$

Finding the scalar product between these new vectors:

$$\begin{aligned}
\left[
\begin{array}{cc}
x_1' \\
y_1' \\
\end{array}
\right] \cdot
\left[
\begin{array}{cc}
x_2' \\
y_2' \\
\end{array}
\right] &= 0 \\
\left(\left[
\begin{array}{cc}
x_1' \\
y_1' \\
\end{array}
\right]\right)^T
\left[
\begin{array}{cc}
x_2' \\
y_2' \\
\end{array}
\right] &= 0 \\
\left(
R \cdot
\left[
\begin{array}{cc}
x_1 \\
y_1 \\
\end{array}
\right]
\right)^T
\left(
R \cdot \left[
\begin{array}{cc}
x_2 \\
y_2 \\
\end{array}
\right]
\right) &= 0 \\
\left(
\left[\begin{array}{cc}
x_1 \\
y_1 \\
\end{array}
\right]^T R^T
\right)
\left(
R \cdot \begin{array}{cc}
x_2 \\
y_2 \\
\end{array}
\right) &= 0 \\
\left[\begin{array}{cc}
x_1 & y_1
\end{array}
\right] R^T R \left[
\begin{array}{cc}
x_2 \\
y_2 \\
\end{array}
\right] &= 0 \\
\left[\begin{array}{cc}
x_1 & y_1
\end{array}
\right] (I) \left[
\begin{array}{cc}
x_2 \\
y_2 \\
\end{array}
\right] &= 0 \\
\left[\begin{array}{cc}
x_1 & y_1
\end{array}
\right]
\left[
\begin{array}{cc}
x_2 \\
y_2 \\
\end{array}
\right] &= 0 \\
\end{aligned}
$$

This shows that even after transformation from a roto-reflective transformation, orthogonal lines are preserved.

For part b)

Let $X$ and $Y$ be points in $R^2$.

The euclidean distance between them gives the distance squared between them as the sum of the squares of the distances along the two axes, that is:

$$\begin{aligned}
d(X, Y)^2 &= (X_1 - Y_1)^2 + (X_2 - Y_2)^2 \\
          &= (X - Y) \cdot (X - Y) \\
          &= (X - Y)^T (X - Y) \\
\end{aligned}$$

In other words, it is the scalar product between the vector formed by subtracting the points and itself.

Applying the transformation $R$ to it:

$$\begin{aligned}
d(R(X), R(Y))^2 &= (R(X) - R(Y)) \cdot (R(X) - R(Y)) \\
                &= R(X - Y) \cdot R(X - Y) \\
                &= (R(X - Y))^T R(X - Y) \\
                &= ((X - Y)^T R^T) R(X - Y) \\
                &= (X - Y)^T (R^T R) (X - Y) \\
                &= (X - Y)^T (I) (X - Y) \\
                &= (X - Y)^T (X - Y) \\
\end{aligned}$$

Since $d(X, Y)^2$ is equal to $d(R(X), R(Y))^2$ distance between points are preserved even after transformation from a roto-reflective transformation

### Complete implementation of function $estimate$ of class $LeastSquareLine$ below. It should update line parameters $a$ and $b$ correpsonding to line model $y=ax+b$. You can use either SVD of matrix $A$ or inverse of matrix $A^T A$, as mentioned in class. NOTE: several cells below test your code.

```python
%matplotlib notebook

import numpy as np
import numpy.linalg as la
import matplotlib
import matplotlib.pyplot as plt
from skimage.measure import ransac
```

```python
class LeastSquareLine:

    def __init__(self):
        self.a = 0.0
        self.b = 0.0
        
    def estimate(self, points2D):
        B = points2D[:,1]
        A = np.copy(points2D)
        A[:,1] = 1.0 

        # Vector B and matrix A are already defined. Change code below
        x = np.linalg.pinv(A.T @ A) @ A.T @ B
#         u, s, vh = np.linalg.svd(A, full_matrices=False)
#         x = vh @ np.linalg.pinv(np.diag(s)) @ u.T @ B
        
        self.a = x[0]
        self.b = x[1]
        return True
        
    def predict(self, x): return (self.a * x) + self.b
    
    def predict_y(self, x): return (self.a * x) + self.b
            
    def residuals(self, points2D):
        return points2D[:,1] - self.predict(points2D[:,0])
    
    def line_par(self):
        return self.a, self.b
```

### (part a) Assume that a set of $N=100$ points in $2D$ includes $N_i=20$ inliers for one line and $N_o=80$ outliers. What is the least number of times one should sample a random pair of points from the set to get probability $p\geq 0.95$ that in at least one of the sampled pairs both points are inliers? Derive a general formula and compute a numerical answer for the specified numbers.

Solution:

Let $I$ be "Inliers" and $O$ be outliers and $n$ be least number of trials one should randomly sample a subset of size 2 such that both samples are inliers.

$$\begin{aligned}
P(\text{at least 1 trial with 2 I in n trials}) &= 1 - P(\text{never get 1 trial with 2 I in n trials}) \\
                                           &= 1 - (P(\text{do NOT get 2I}))^n \\
                                           &= 1 - (1 - P(\text{DO get 2I}))^n \\
                                           &= 1 - \left(1 - \frac{{I \choose 2}}{{I \choose 2} + {O \choose 2} + {O \choose 1} \cdot {I \choose 1}}\right)^n \\
                                           &>= 0.95
\end{aligned}$$

Simplifying leads to the following:

$$\begin{aligned}
1 - \left(1 - \frac{{I \choose 2}}{{I \choose 2} + {O \choose 2} + {O \choose 1} \cdot {I \choose 1}}\right)^n &>= 0.95 \\
\left(1 - \frac{{I \choose 2}}{{I \choose 2} + {O \choose 2} + {O \choose 1} \cdot {I \choose 1}}\right)^n &<= 0.05 \\
\left(1 - \frac{\frac{I (I - 1)}{2}}{\frac{I (I - 1)}{2} + \frac{O (O - 1)}{2} + IO}\right)^n &<= 0.05 \\
\left(1 - \frac{I(I - 1)}{I(I - 1) + O(O - 1) + 2IO}\right)^n &<= 0.05 \\
\end{aligned}$$

Let $a = P\left(\text{DO get 2I}\right) = \frac{I(I - 1)}{I(I - 1) + O(O - 1) + 2IO}$.

The equation becomes:

$$\begin{aligned}
\left(1 - \frac{I(I - 1)}{I(I - 1) + O(O - 1) + 2IO}\right)^n &<= 0.05 \\
\left(1 - a\right)^n &<= 0.05 \\
log_\left(1 - a\right) \left(1- a\right)^n &>= log_\left(1 - a\right) 0.05 \\
n &>= log_\left(1 - a\right) 0.05 \\
n &>= \frac{\log 0.05}{\log \left(1 - a\right)} \\
n &>= \frac{\log 0.05}{\log \left(1 - \frac{I(I - 1)}{I(I - 1) + O(O - 1) + 2IO}\right)} \\
\end{aligned}$$

Solving for $I = 20$ and $O = 80$ as in the question:

$$\begin{aligned}
n &>= \frac{\log 0.05}{\log \left(1 - \frac{I(I - 1)}{I(I - 1) + O(O - 1) + 2IO}\right)} \\
n &>= \frac{\log 0.05}{\log \left(1 - \frac{(20)((20) - 1)}{(20)((20) - 1) + (80)((80) - 1) + 2(20)(80)}\right)} \\
n &>= \frac{\log 0.05}{\log \left(1 - \frac{380}{9,520}\right)} \\
n &>\approx 73.54
\end{aligned}$$

Therefore 74 is the minimum number of samples $n$ that we must make to have a probability $p >= 0.95$ that at least one of the sampled pairs both have inliers.

```python
def draw_k_lines(K, data):
    data_coords = np.roll(
        np.argwhere(data), 1
    )
    line_x = np.arange(0, data.shape[1])
    for i in range(K):
        inliers_k = np.array([0])
        while (
            inliers_k is None
            or np.sum(inliers_k)
            < -4 * K + 160
        ):
            (
                model_robust_k,
                inliers_k,
            ) = ransac(
                data_coords,
                LeastSquareLine,
                min_samples=2,
                residual_threshold=1,
                max_trials=200,
            )
            print(
                "Found ",
                0
                if inliers_k is None
                else np.sum(inliers_k),
                " inliers.",
                end="\r",
            )
        print("")
        old_data_coords_inliers = (
            data_coords.shape[0]
        )
        data_coords = np.delete(
            data_coords,
            inliers_k,
            axis=0,
        )
        print(
            "Removed ",
            old_data_coords_inliers
            - data_coords.shape[0],
            " inliers.",
        )
        (
            a_rs_k,
            b_rs_k,
        ) = model_robust_k.line_par()
        print(
            f"Formed this line #{i + 1}: y = {a_rs_k}x + {b_rs_k}"
        )

        # generate coordinates of estimated models
        line_y_robust_k = (
            model_robust_k.predict_y(
                line_x
            )
        )
        in_range_indices = (
            line_y_robust_k
            < data.shape[0]
        ) * (line_y_robust_k >= 0)
        x_vals = line_x[
            in_range_indices
        ]
        y_vals = line_y_robust_k[
            in_range_indices
        ]
        ax.plot(
            x_vals,
            y_vals,
            label="RANSAC line {} (a={:4.2f}, b={:4.1f})".format(
                i, a_rs_k, b_rs_k
            ),
        )
    plt.show()
    print(f"Done finding {K} lines.")


draw_k_lines(K, can)
```
